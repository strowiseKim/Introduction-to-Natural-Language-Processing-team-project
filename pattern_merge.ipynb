```python
import pandas as pd
import numpy as np
name = ['cft', 'dpo', 'chosen', 'rejected', 'human']

data = []

for x in range(5):
    data.append(pd.read_csv('data_'+name[x]+'.csv'))

train = pd.read_csv('mtld_results.csv')
essay = pd.read_csv('essays_mtld_results.csv')

data[2]['Measure of Textual Lexical Diversity'] = train['chosen_mtld']
data[3]['Measure of Textual Lexical Diversity'] = train['rejected_mtld']

data[0]['Measure of Textual Lexical Diversity'] = essay['CFT_mtld']
data[1]['Measure of Textual Lexical Diversity'] = essay['DPO_mtld']
data[4]['Measure of Textual Lexical Diversity'] = essay['Human_mtld']

for x in range(5):
    data[x].to_csv('final_data_'+name[x]+'.csv', index = False)
```

```python
mean = []

for x in range(4):
    mean.append(data[x].mean())

delta_train = mean[2] - mean[3]
delta_essay = mean[1] - mean[0]

df_delta = pd.DataFrame({
    "delta_train": delta_train,
    "delta_essay": delta_essay
})

df_delta = df_delta.reset_index().rename(columns={"index": "feature_name"})

df_delta.to_csv('delta.csv', index = False)
```
